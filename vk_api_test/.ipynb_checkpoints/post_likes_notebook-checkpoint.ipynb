```python
import requests
import json

result = requests.get("https://api.vk.com/method/wall.get?domain=roctbb&count=1&v=5.60")
json_result = json.loads(result.text)

posts = []
col = json_result["response"]["count"]

for offset in range(0,col,100):
    result = requests.get(
        "https://api.vk.com/method/wall.get?domain=roctbb&count=100&offset="+str(offset)+"&v=5.60")
    print(offset)
    json_result = json.loads(result.text)
    posts = posts + json_result["response"]["items"]
```

```python
for post in posts:
    print(post["text"]+ " - " + str(post["likes"]["count"]))
```
